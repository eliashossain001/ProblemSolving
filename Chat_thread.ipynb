```python
import random
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

class ChatApp:
    def __init__(self):
        self.all_conversations = []

    def generate_response(self, message):
        responses = [
            "Hello!",
            "How are you?",
            "Nice to meet you!",
            "What can I do for you?",
            "I'm sorry, I don't understand.",
            "Tell me more.",
            "That's interesting.",
            "Goodbye!"
        ]
        return random.choice(responses)

    def start_chat(self):
        conversation = []
        print("Start chatting. Type 'exit' to end the conversation.")

        while True:
            user_message = input("You: ")

            if user_message.lower() == 'exit':
                break

            bot_response = self.generate_response(user_message)
            conversation.append({"User": user_message, "Bot": bot_response})

            print(f"Bot: {bot_response}")

        print("Conversation ended.")
        self.all_conversations.append(conversation)

    def generate_title(self, conversation):
        if conversation:
            return ' '.join(conversation[0]["User"].split()[:3])
        else:
            return "Untitled Conversation"

    def show_conversations(self):
        for index, conversation in enumerate(self.all_conversations):
            title = self.generate_title(conversation)
            print(f"{index}: {title}")

    def show_conversation(self, index):
        if 0 <= index < len(self.all_conversations):
            conversation = self.all_conversations[index]
            title = self.generate_title(conversation)
            print(f"\n{title} - Conversation:")
            for item in conversation:
                print(f"You: {item['User']}")
                print(f"Bot: {item['Bot']}\n")
        else:
            print("Invalid conversation index.")

def main():
    chat_app = ChatApp()

    while True:
        chat_app.start_chat()

        ask_for_another_chat = input("Do you want to start another chat? (yes/no): ")
        if ask_for_another_chat.lower() != 'yes':
            break

    chat_app.show_conversations()

    @interact
    def select_conversation(conversation_index=list(range(len(chat_app.all_conversations)))):
        chat_app.show_conversation(conversation_index)

if __name__ == "__main__":
    main()
```

```python
import random
from datetime import datetime, timedelta

def generate_response(message):
    responses = [
        "Hello!",
        "How are you?",
        "Nice to meet you!",
        "What can I do for you?",
        "I'm sorry, I don't understand.",
        "Tell me more.",
        "That's interesting.",
        "Goodbye!"
    ]
    return random.choice(responses)

def generate_title(conversation):
    if conversation:
        # Get the timestamp of the first message in the conversation
        timestamp = conversation[0].get("Timestamp", None)

        if timestamp:
            # Get the current date and yesterday's date
            today = datetime.now().date()
            yesterday = today - timedelta(days=1)

            # Check if the conversation happened today or yesterday
            if timestamp.date() == today:
                extra_title = "Today's Chat"
            elif timestamp.date() == yesterday:
                extra_title = "Yesterday's Chat"
            else:
                extra_title = None

            # Keep the existing title
            existing_title = ' '.join(conversation[0]["User"].split()[:3]) if conversation[0].get("Title") else None

            # Combine existing title and extra title
            if existing_title and extra_title:
                return f"{existing_title} ({extra_title})"
            elif existing_title:
                return existing_title
            elif extra_title:
                return extra_title

    return "Untitled Conversation"

def print_conversation(conversation):
    for item in conversation:
        print(f"You: {item['User']}")
        print(f"Bot: {item['Bot']}")
        print()

def main():
    all_conversations = []

    while True:
        conversation = []
        print("Start chatting. Type 'exit' to end the conversation.")

        while True:
            user_message = input("You: ")

            if user_message.lower() == 'exit':
                break

            bot_response = generate_response(user_message)
            conversation.append({"User": user_message, "Bot": bot_response, "Timestamp": datetime.now()})

            print(f"Bot: {bot_response}")

        print("Conversation ended.")
        # Set the title for the conversation
        conversation[0]["Title"] = generate_title(conversation)
        all_conversations.append(conversation)

        ask_for_another_chat = input("Do you want to start another chat? (yes/no): ")
        if ask_for_another_chat.lower() != 'yes':
            break

    print("\nAll Conversations:")
    for index, conversation in enumerate(all_conversations, 1):
        title = generate_title(conversation)
        print(f"\nConversation {index} - Title: {title}:")
        print_conversation(conversation)

if __name__ == "__main__":
    main()
```
